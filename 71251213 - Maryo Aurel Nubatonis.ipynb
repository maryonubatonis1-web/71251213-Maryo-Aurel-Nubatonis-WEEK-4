```python
try:
    suhu = int(input("Masukkan suhu tubuh: "))
    if suhu >= 38:
        print("Anda demam")
    else:
        print("Anda tidak demam")
except:
    print("Input tidak valid! Harap masukkan angka untuk suhu tubuh.")
```

```python
try:
    bilangan = int(input("Masukkan suatu bilangan: "))
    if bilangan > 0:
        print("Positif")
    elif bilangan < 0:
        print("Negatif")
    elif bilangan == 0:
        print("Nol")
except:
    print("Input tidak valid! Harap masukkan angka bulat.")
```

```python
try:
    a = int(input("Masukkan bilangan pertama: "))
    b = int(input("Masukkan bilangan kedua: "))
    c = int(input("Masukkan bilangan ketiga: "))

    if a > b and a > c:
        print("Terbesar: ", a)
    elif b > a and b > c:
        print("Terbesar: ", b)
    elif c > a and c > b:
        print("Terbesar: ", c)
except:
    print("Terjadi kesalahan! Pastikan semua input adalah angka.")
```

```python
try:
    bilangan = int(input("Masukkan suatu bilangan: "))
    
    hasil = "Positif" if bilangan > 0 else ("Negatif" if bilangan < 0 else "Nol")
    
    print(hasil)
except:
    print("Input tidak valid! Harap masukkan angka.")
```

```python
try:
    bulan = int(input("Masukkan bulan (1-12): "))

    if bulan == 2:
        print("Jumlah hari: 29")
    elif bulan in [4, 6, 9, 11]:
        print("Jumlah hari: 30")
    elif bulan in [1, 3, 5, 7, 8, 10, 12]:
        print("Jumlah hari: 31")
    else:
        print("Pesan: Bulan yang diinputkan tidak valid")

except:
    print("Pesan: Input tidak valid, harap masukkan angka 1-12")
```

```python
try:
    sisi1 = int(input("Masukkan sisi 1: "))
    sisi2 = int(input("Masukkan sisi 2: "))
    sisi3 = int(input("Masukkan sisi 3: "))

    if sisi1 == sisi2 and sisi2 == sisi3:
        print("3 sisi sama")
    
    elif sisi1 == sisi2 or sisi1 == sisi3 or sisi2 == sisi3:
        print("2 sisi sama")
    
    else:
        print("Tidak ada yang sama")

except:
    print("Input tidak valid! Harap masukkan angka yang benar.")
```
